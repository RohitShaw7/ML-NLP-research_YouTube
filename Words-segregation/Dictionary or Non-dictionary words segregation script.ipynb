```python
import numpy as np
import pandas as pd
import enchant
import re
```

```python
df=pd.read_csv('data_cleaned_V3.csv',delimiter=',')
df1=df.dropna()
```

```python
from nltk.corpus import wordnet
nonDict=[]
Dict=[]
test=df1['comment']
for i in test:
    res = re.findall(r'\w+', i)
    for j in res:
        if not wordnet.synsets(j):
          nonDict.append(j)
        else:
          Dict.append(j)
d = enchant.Dict("en_US")
for j in nonDict:
    if(d.check(j)):
        Dict.append(j)
        nonDict= [i for i in nonDict if i != j]
```

```python
print(sorted(set(Dict)))
```
