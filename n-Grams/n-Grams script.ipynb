# Calculating n-Grams

```python
import pandas as pd
import numpy as np
import nltk
from nltk.util import ngrams
```

```python
def extract_ngrams(data, num):
    n_grams = ngrams(nltk.word_tokenize(data), num)
    return [ ' '.join(grams) for grams in n_grams]
```

```python
data = pd.read_csv('this_is_america_cleaned_V1.csv',delimiter=",")

copy_data = data['COMMENT'].dropna()

for i in copy_data[0:199]:      # I have done it for first 200 sentences
    print("Sentence: ",i)       # Prints each sentence
    print("1-gram: ", extract_ngrams(i,1))       # Prints 1-gram
    print("2-gram: ", extract_ngrams(i, 2))      # Prints 2-gram
    print("3-gram: ", extract_ngrams(i, 3))      # Prints 3-gram and similarly we can do it for n(user defined) grams
```
